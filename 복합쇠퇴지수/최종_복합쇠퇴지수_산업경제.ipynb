# 2. 데이터 전처리

## 1) 총종사자 증감률

```python
## 데이터프레임의 규칙을 보면, 컬럼이 '단순연도' 인 경우(예.2013, 2014) '총 종사자수'를 의미함. 

# 동이름 컬럼과 함께'총 종사자수' 내용을 담은 새 데이터프레임 생성
employee = df1[['동별(1)','2013','2014','2015','2016','2017','2018']] 

# 불필요한 0행 삭제 
employee.drop(0, inplace=True) 

# 행 삭제에 따른 인덱스 재설정 
employee.reset_index(drop = True, inplace = True) 

# 컬럼명 변경 
employee.rename(columns={'동별(1)':'dong'}, inplace=True) 

employee
```

```python
employee.info()
```

```python
#'송도1동 + 송도2동 + 송도3동 = 송도동'으로 계산해서 세 동을 합친 '송도동' 수준에서 보고자 함. (법정동별 구분)

# 위와 같은 배경에서, null 값을 0으로 변환
employee.iloc[13,1] = 0

# object -> int64 형변환 
employee = employee.astype({"2013":"int64", "2014":"int64", "2015":"int64", "2016":"int64", "2017":"int64", "2018":"int64"})
employee.info()
employee
```

```python
# '송'으로 시작하는 값을 regex에 할당함 
regex = '송.*'

# 새로운 14행을 생성 = 송도1동 + 송도2동 + 송도3동. 다만, 'dong'열은 제외하고 합산하기. 따라서, 14행 'dong' 열은 null값.
employee.loc[14,1:] = employee[employee.dong.str.match(regex)].sum()

# null 값이 들어 있는 14행 'dong'열에 '송도동' 명칭 부여. 
employee['dong'].replace({None: '송도동'}, inplace=True)
employee
```

```python
# 옥련1동 + 옥련2동 = 옥련동
# 연수1동 + 연수2동 + 연수3동 = 연수동
# 동춘1동 + 동춘2동 + 동춘3동 = 동춘동

okryun = employee.iloc[1:3, 1:].sum()
yeonsu = employee.iloc[4:7, 1:].sum()
dongchun = employee.iloc[8:11, 1:].sum()

employee.loc[15, 1:] = okryun
employee.iloc[15,0] = '옥련동'

employee.loc[16, 1:] = yeonsu
employee.iloc[16,0] = '연수동'

employee.loc[17, 1:] = dongchun
employee.iloc[17,0] = '동춘동'

employee
```

```python
# 필요한 행만 추출하여 데이터프레임 재생성

filter = ['옥련동', '선학동', '연수동', '청학동', '동춘동', '송도동']

employee2 = employee.query("dong == @filter")
employee2.reset_index(drop=True, inplace=True)
employee2
```

```python
# (기준연도 총종사자수 - 비교연도 총종사자수) / 비교연도 총종사자수 * 100
# 기준연도: 2018, 비교연도: 2013
# 산출 근거: 국토교통부(2021, p.12)

employee2['employee_chg_rate(13-18)'] = (employee2['2018'] - employee2['2013'])/employee2['2013']*100
employee2
```

```python
# 상대감소지수 변환 (최대값 – 원래값) / (최대값 – 최소값) × 100

employee3 = employee2.copy()

max = employee3['employee_chg_rate(13-18)'].max()
min = employee3['employee_chg_rate(13-18)'].min()

employee3['employee_decr_idx(13-18)'] = (max -employee3['employee_chg_rate(13-18)']) / (max - min) *100
employee3

# 지수 표준화 (쇠퇴진단지표 - 하한값) / (상한값 - 하한값)

max = employee3['employee_decr_idx(13-18)'].max()
min = employee3['employee_decr_idx(13-18)'].min()

employee3['std_employee_decr_idx(13-18)'] = (employee3['employee_decr_idx(13-18)'] - min) / (max - min) 
employee3
```

```python
# 태블로작업을 위한 데이터 프레임 (연도별 총 종사자수)

employee4 = employee3.iloc[:, :7]

tableau1 = pd.melt(employee4,
       id_vars = 'dong',
       var_name = 'year',
       value_name = 'employee').sort_values(by=['dong', 'year']).reset_index(drop=True)

tableau2 =  employee3.iloc[:, [0,7,8,9]]
```

```python
# 데이터 내보내기
employee3.to_csv('최종_변환_산업경제_총종사자증감률_2013-2018.csv')
tableau1.to_csv('태블로_산업경제_총종사자수_2013-2018.csv')
tableau2.to_csv('태블로_산업경제_총종사자증감률_2013-2018.csv')
```

```python
## 법정동별 구분 

# 동춘1동 + 동춘2동 + 동춘3동 = 동춘동
# 송도1동 + 송도2동 + 송도3동 = 송도동
# 연수1동 + 연수2동 + 연수3동 = 연수동
# 옥련1동 + 옥련2동 = 옥련동

okryun = 제조업체증감율.iloc[1:3, 1:].sum()
yeonsu = 제조업체증감율.iloc[4:7, 1:].sum()
dongchun = 제조업체증감율.iloc[8:11, 1:].sum()
songdo = 제조업체증감율.iloc[11:14, 1:].sum()

제조업체증감율.loc[14, 1:] = okryun
제조업체증감율.iloc[14,0] = '옥련동'

제조업체증감율.loc[15, 1:] = yeonsu
제조업체증감율.iloc[15,0] = '연수동'

제조업체증감율.loc[16, 1:] = dongchun
제조업체증감율.iloc[16,0] = '동춘동'


제조업체증감율.loc[17, 1:] = songdo
제조업체증감율.iloc[17,0] = '송도동'

제조업체증감율
```

```python
# 필요힌 행만 추출하여 데이터프레임 재생성

filter = ['옥련동', '선학동', '연수동', '청학동', '동춘동', '송도동']

제조업체증감율 = 제조업체증감율.query("dong == @filter")
제조업체증감율.reset_index(drop=True, inplace=True)
제조업체증감율
```

```python
제조업체증감율= 제조업체증감율.sort_values(by=['dong']).reset_index(drop=True)

# (기준연도 총제조업체수 - 비교연도 총제조업체수) / 비교연도 총제조업체수 * 100
# 기준연도: 2018, 비교연도: 2013

제조업체증감율['증감율 2013-2018'] = (제조업체증감율['2018'] - 제조업체증감율['2013'] ) / 제조업체증감율['2013'] * 100
제조업체증감율
```

```python
# 태블로 시각화 작업을 위해 데이터프레임 구조 변경

제조업체수_tableau1 = 제조업체증감율[['dong','2013','2014','2015','2016','2017','2018']]


제조업체수_tableau1 = pd.melt(제조업체수_tableau1,
               id_vars = 'dong',
               var_name = 'year',
               value_name = '제조업체수')

제조업체수_tableau1
```

```python
# 재정렬

제조업체수_tableau1= 제조업체수_tableau1.sort_values(by=['dong','year'])
제조업체수_tableau1.reset_index(drop=True,inplace=True)
제조업체수_tableau1
```

```python
# 상대감소지수 (최대값 – 원래값) / (최대값 – 최소값) × 100

max1 = 제조업체증감율['증감율 2013-2018'].max()
min1 = 제조업체증감율['증감율 2013-2018'].min()

제조업체증감율['상대감소지수'] = (max1-제조업체증감율['증감율 2013-2018']) / (max1 - min1) * 100

# 표준화 (쇠퇴진단지표 - 하한값) / (상한값 - 하한값)

max2 = 제조업체증감율['상대감소지수'].max()
min2 = 제조업체증감율['상대감소지수'].min()

제조업체증감율['std'] = (제조업체증감율['상대감소지수'] - min2) / (max2 - min2)

제조업체증감율
```

```python
제조업체수_tableau1.to_csv('태블로_산업경제_제조업체증감율_제조업체수.csv')
제조업체증감율.to_csv('최종_변환_산업경제_제조업체증감율_2013-2018.csv')
```

```python
## 법정동별
# 동춘1동 + 동춘2동 + 동춘3동  = 동춘동
# 송도1동 + 송도2동 + 송도3동 = 송도동
# 연수1동 + 연수2동 + 연수3동 = 연수동
# 옥련1동 + 옥련2동 = 옥련동

okryun = 도소매업체증감율.iloc[1:3, 1:].sum()
yeonsu = 도소매업체증감율.iloc[4:7, 1:].sum()
dongchun = 도소매업체증감율.iloc[8:11, 1:].sum()
songdo = 도소매업체증감율.iloc[11:14, 1:].sum()

도소매업체증감율.loc[14, 1:] = okryun
도소매업체증감율.iloc[14,0] = '옥련동'

도소매업체증감율.loc[15, 1:] = yeonsu
도소매업체증감율.iloc[15,0] = '연수동'

도소매업체증감율.loc[16, 1:] = dongchun
도소매업체증감율.iloc[16,0] = '동춘동'


도소매업체증감율.loc[17, 1:] = songdo
도소매업체증감율.iloc[17,0] = '송도동'

도소매업체증감율
```

```python
# 필요한 행만 추출하여 '동, 연도별 총 인구수' 데이터프레임 생성

filter = ['옥련동', '선학동', '연수동', '청학동', '동춘동', '송도동']

도소매업체증감율 = 도소매업체증감율.query("dong == @filter")
도소매업체증감율.reset_index(drop=True, inplace=True)
도소매업체증감율
```

```python
# (기준연도 총도소매업체수 - 비교연도 총도소매업체수) / 비교연도 총도소매업업체수 * 100
# 기준연도: 2018, 비교연도: 2013

도소매업체증감율= 도소매업체증감율.sort_values(by=['dong']).reset_index(drop=True)
도소매업체증감율['증감율 2013-2018'] = (도소매업체증감율['2018'] - 도소매업체증감율['2013'] ) / 도소매업체증감율['2013'] * 100
도소매업체증감율
```

```python
# 태블로 시각화 작업 위해 데이터프레임 재구조화

도소매업체수_tableau1 = 도소매업체증감율[['dong','2013','2014','2015','2016','2017','2018']]

도소매업체수_tableau1 = pd.melt(도소매업체수_tableau1,
               id_vars = 'dong',
               var_name = 'year',
               value_name = '도소매업체수')

# 재정렬
도소매업체수_tableau1=도소매업체수_tableau1.sort_values(by=['dong','year'])
도소매업체수_tableau1.reset_index(drop=True, inplace=True)

도소매업체수_tableau1
```

```python
# 상대감소지수 (최대값 – 원래값) / (최대값 – 최소값) × 100

max1 = 도소매업체증감율['증감율 2013-2018'].max()
min1 = 도소매업체증감율['증감율 2013-2018'].min()

도소매업체증감율['상대감소지수'] = (max1-도소매업체증감율['증감율 2013-2018']) / (max1 - min1) * 100

# 표준화 (쇠퇴진단지표 - 하한값) / (상한값 - 하한값)

max2 = 도소매업체증감율['상대감소지수'].max()
min2 = 도소매업체증감율['상대감소지수'].min()

도소매업체증감율['std'] = (도소매업체증감율['상대감소지수'] - min2) / (max2 - min2)

도소매업체증감율
```

```python
# 고차산업종사자 비율

highlevel2['h_rate_2013'] = (highlevel2['h_2013'] / highlevel2['2013']) *100
highlevel2['h_rate_2014'] = (highlevel2['h_2014'] / highlevel2['2014']) *100
highlevel2['h_rate_2015'] = (highlevel2['h_2015'] / highlevel2['2015']) *100
highlevel2['h_rate_2016'] = (highlevel2['h_2016'] / highlevel2['2016']) *100
highlevel2['h_rate_2017'] = (highlevel2['h_2017'] / highlevel2['2017']) *100
highlevel2['h_rate_2018'] = (highlevel2['h_2018'] / highlevel2['2018']) *100

# 구성비치환 변환 100 - 원래값

highlevel2['non_h_rate_2013'] = 100 - highlevel2['h_rate_2013']
highlevel2['non_h_rate_2014'] = 100 - highlevel2['h_rate_2014']
highlevel2['non_h_rate_2015'] = 100 - highlevel2['h_rate_2015']
highlevel2['non_h_rate_2016'] = 100 - highlevel2['h_rate_2016']
highlevel2['non_h_rate_2017'] = 100 - highlevel2['h_rate_2017']
highlevel2['non_h_rate_2018'] = 100 - highlevel2['h_rate_2018']

# 지수 표준화 (쇠퇴진단지표 - 하한값) / (상한값 - 하한값)

highlevel3 = highlevel2.copy()

highlevel3['std_non_h_rate_2013'] = (highlevel3['non_h_rate_2013'] - highlevel3['non_h_rate_2013'].min()) / (highlevel3['non_h_rate_2013'].max() - highlevel3['non_h_rate_2013'].min()) 
highlevel3['std_non_h_rate_2014'] = (highlevel3['non_h_rate_2014'] - highlevel3['non_h_rate_2014'].min()) / (highlevel3['non_h_rate_2014'].max() - highlevel3['non_h_rate_2014'].min())
highlevel3['std_non_h_rate_2015'] = (highlevel3['non_h_rate_2015'] - highlevel3['non_h_rate_2015'].min()) / (highlevel3['non_h_rate_2015'].max() - highlevel3['non_h_rate_2015'].min()) 
highlevel3['std_non_h_rate_2016'] = (highlevel3['non_h_rate_2016'] - highlevel3['non_h_rate_2016'].min()) / (highlevel3['non_h_rate_2016'].max() - highlevel3['non_h_rate_2016'].min()) 
highlevel3['std_non_h_rate_2017'] = (highlevel3['non_h_rate_2017'] - highlevel3['non_h_rate_2017'].min()) / (highlevel3['non_h_rate_2017'].max() - highlevel3['non_h_rate_2017'].min()) 
highlevel3['std_non_h_rate_2018'] = (highlevel3['non_h_rate_2018'] - highlevel3['non_h_rate_2018'].min()) / (highlevel3['non_h_rate_2018'].max() - highlevel3['non_h_rate_2018'].min()) 

highlevel3
```
