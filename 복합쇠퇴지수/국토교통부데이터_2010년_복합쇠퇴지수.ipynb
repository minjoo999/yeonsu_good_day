```python
# 불필요한 컬럼 삭제

# 인구증감률과 인구변화율은 동일한 지표임. 
# 인구총조사가 아닌 주민등록인구통계 기준 인구변화율을 선택. 

incheon.drop(['시군구명','인구변화율(인구총조사)', '노령화지수(인구총조사)','소형주택비율','고령인구수(인구총조사)', '고령인구수(주민등록인구통계)', 
              '노년부양비(인구총조사)', '노년부양비(주민등록인구통계)', '접도율', '노후건축물비율',
              '시군구코드', '읍면동코드', '사업체당종사자수', '총사업체수', 
              '경제활동인구(인구총조사)', '경제활동인구(주민등록인구통계)','주요산업별종사자수',
              '제조업종사자비율','도소매업종사자수(천명당)'
              ],
             axis=1, inplace=True)
incheon
```

```python
# null 값 없는 행만 추출했을 때, 연도 구축이 어떻게 되어있는지 확인하기 
incheon.dropna(axis=0)['연도'].unique()
```

```python
df = incheon.copy()
```

```python
# null 값 포함행 삭제 
df= df.dropna(axis=0)
df
```

```python
df.info()
```

```python
# 연수구의 동만 추출하여 새로운 데이터 프레임 생성

new = df[df['읍면동명'].str.contains('연수|청학|선학|동춘|송도|옥련')]
new.reset_index(drop=True, inplace=True)
new
```

```python
del new['연도']
new
```

```python
# '연'으로 시작하는 값을 regex에 할당함 
regex = '연.*'

# 새로운 7행을 생성 = 연수1동 + 연수2동. 다만, '읍면동명'열은 제외하고 합산하기. 따라서, 17행 '읍면동명' 열은 null값.
new.loc[7,1:] = new[new.읍면동명.str.match(regex)].mean()

# null 값이 들어 있는 14행 '읍면동'열에 '연수동' 명칭 부여. 
new['읍면동명'].replace({None: '연수동'}, inplace=True)
new
```

```python
# 위와 동일 작업 반복 

regex = '옥.*'

new.loc[8,1:] = new[new.읍면동명.str.match(regex)].mean()

new['읍면동명'].replace({None: '옥련동'}, inplace=True)
new
```

```python
# 필요한 행만 추출하여 데이터프레임 재생성

filter = ['옥련동', '선학동', '연수동', '청학동', '동춘1동']

new2 = new.query("읍면동명 == @filter")
new2.reset_index(drop=True, inplace=True)
new2
```

```python
#동춘1동을 동춘동으로 변경
new2.iloc[0,0] = '동춘동'
new2
```

```python
df = new2.copy()
```

```python
# 구성비치환 작업 (100-원래값)

df['신규주택비율'] = 100 - df['신규주택비율']
df['고차산업종사자비율'] = 100 - df['고차산업종사자비율']
```

```python
# 상대감소지수 (최대값 – 원래값) / (최대값 – 최소값) × 100

df['인구변화율(주민등록인구통계)'] = (df['인구변화율(주민등록인구통계)'].max() - df['인구변화율(주민등록인구통계)']) / (df['인구변화율(주민등록인구통계)'].max() - df['인구변화율(주민등록인구통계)'].min())
df['노령화지수(주민등록인구통계)'] = (df['노령화지수(주민등록인구통계)'].max() - df['노령화지수(주민등록인구통계)']) / (df['노령화지수(주민등록인구통계)'].max() - df['노령화지수(주민등록인구통계)'].min())
df['총사업체수증감율'] = (df['총사업체수증감율'].max() - df['총사업체수증감율']) / (df['총사업체수증감율'].max() - df['총사업체수증감율'].min())
df['총종사자수증감율'] = (df['총종사자수증감율'].max() - df['총종사자수증감율']) / (df['총종사자수증감율'].max() - df['총종사자수증감율'].min())
df['도소매업체증감율'] = (df['도소매업체증감율'].max() - df['도소매업체증감율']) / (df['도소매업체증감율'].max() - df['도소매업체증감율'].min())
df['제조업체증감율'] = (df['제조업체증감율'].max() - df['제조업체증감율']) / (df['제조업체증감율'].max() - df['제조업체증감율'].min())
```

```python
# 역입지계수 (최대값 – 원래값) / 최대값 × 100

df['종사자수(천명당)'] = (df['종사자수(천명당)'].max() - df['종사자수(천명당)']) / df['종사자수(천명당)'].max() * 100
df['부동산공시지가'] = (df['부동산공시지가'].max() - df['부동산공시지가']) / df['부동산공시지가'].max() * 100
```

```python
# 표준화 작업 (쇠퇴진단지표 - 하한값) / (상한값 - 하한값)

df['신규주택비율'] = (df['신규주택비율'] - df['신규주택비율'].min()) / (df['신규주택비율'].max() - df['신규주택비율'].min())
df['고차산업종사자비율'] = (df['고차산업종사자비율'] - df['고차산업종사자비율'].min()) / (df['고차산업종사자비율'].max() - df['고차산업종사자비율'].min())
df['인구변화율(주민등록인구통계)'] = (df['인구변화율(주민등록인구통계)'] - df['인구변화율(주민등록인구통계)'].min()) / (df['인구변화율(주민등록인구통계)'].max() - df['인구변화율(주민등록인구통계)'].min())
df['노령화지수(주민등록인구통계)'] = (df['노령화지수(주민등록인구통계)'] - df['노령화지수(주민등록인구통계)'].min()) / (df['노령화지수(주민등록인구통계)'].max() - df['노령화지수(주민등록인구통계)'].min())
df['총사업체수증감율'] = (df['총사업체수증감율'] - df['총사업체수증감율'].min()) / (df['총사업체수증감율'].max() - df['총사업체수증감율'].min())
df['총종사자수증감율'] = (df['총종사자수증감율'] - df['총종사자수증감율'].min()) / (df['총종사자수증감율'].max() - df['총종사자수증감율'].min())
df['도소매업체증감율'] = (df['도소매업체증감율'] - df['도소매업체증감율'].min()) / (df['도소매업체증감율'].max() - df['도소매업체증감율'].min())
df['제조업체증감율'] = (df['제조업체증감율'] - df['제조업체증감율'].min()) / (df['제조업체증감율'].max() - df['제조업체증감율'].min())
df['종사자수(천명당)'] = (df['종사자수(천명당)'] - df['종사자수(천명당)'].min()) / (df['종사자수(천명당)'].max() - df['종사자수(천명당)'].min())
df['부동산공시지가'] = (df['부동산공시지가'] - df['부동산공시지가'].min()) / (df['부동산공시지가'].max() - df['부동산공시지가'].min())
df['노후주택비율'] = (df['노후주택비율'] - df['노후주택비율'].min()) / (df['노후주택비율'].max() - df['노후주택비율'].min())
df['기초생활수급자(천인당)'] = (df['기초생활수급자(천인당)'] - df['기초생활수급자(천인당)'].min()) / (df['기초생활수급자(천인당)'].max() - df['기초생활수급자(천인당)'].min())
```

```python
# 각 지표의 표준화점수에 가중치를 곱하여 합산 

df_merged2 = df.copy()

df_merged2['인구사회쇠퇴지수'] = (df_merged2['인구변화율(주민등록인구통계)']*9.246193836) + (df_merged2['노령화지수(주민등록인구통계)']*9.110038371) + (df_merged2['기초생활수급자(천인당)']*8.35499443)

df_merged2['산업경제쇠퇴지수'] = (df_merged2['총종사자수증감율']*5.656640673) + (df_merged2['총사업체수증감율']*9.865082312) + (df_merged2['제조업체증감율']*7.389528407) + (df_merged2['도소매업체증감율']*9.097660602) + (df_merged2['종사자수(천명당)']*8.144572348) + (df_merged2['고차산업종사자비율']*6.498329001)

df_merged2['물리환경쇠퇴지수'] = (df_merged2['노후주택비율']*7.637083797) + (df_merged2['신규주택비율']*8.862482981) + (df_merged2['부동산공시지가']*10.13739324)

df_merged2['복합쇠퇴지수'] = (df_merged2['인구변화율(주민등록인구통계)']*9.246193836) + (df_merged2['노령화지수(주민등록인구통계)']*9.110038371) + (df_merged2['기초생활수급자(천인당)']*8.35499443) + (df_merged2['총종사자수증감율']*5.656640673) + (df_merged2['총사업체수증감율']*9.865082312) + (df_merged2['제조업체증감율']*7.389528407) + (df_merged2['도소매업체증감율']*9.097660602) + (df_merged2['종사자수(천명당)']*8.144572348) + (df_merged2['고차산업종사자비율']*6.498329001) + (df_merged2['노후주택비율']*7.637083797) + (df_merged2['신규주택비율']*8.862482981) + (df_merged2['부동산공시지가']*10.13739324)

df_merged2
```

```python
# 지수값에 따른 순위 산출

df_merged3 =  df_merged2.copy()

df_merged3['순위(복합)'] = df_merged3['복합쇠퇴지수'].rank(ascending=False) 
df_merged3['순위(인구사회)'] = df_merged3['인구사회쇠퇴지수'].rank(ascending=False) 
df_merged3['순위(산업경제)'] = df_merged3['산업경제쇠퇴지수'].rank(ascending=False) 
df_merged3['순위(물리환경)'] = df_merged3['물리환경쇠퇴지수'].rank(ascending=False) 

df_merged3
```
