```python
import pandas as pd


# 인구사회 부문
df10 = pd.read_csv('최종_변환_인구사회_인구증감률_2013-2022.csv', index_col = 0)
df13 = pd.read_csv('최종_변환_인구사회_노령화지수_2021.csv', index_col = 0) 
df12 = pd.read_csv('최종_변환_인구사회_천명당기초생활수급자수.csv', index_col = 0)

# 산업경제 부문 
df3 = pd.read_csv('최종_변환_산업경제_도소매업체증감율_2013-2018.csv', index_col = 0)
df5 = pd.read_csv('최종_변환_산업경제_제조업체증감율_2013-2018.csv',index_col=0)
df7 = pd.read_csv('최종_변환_산업경제_천명당총종사자수_2013-2018.csv', index_col = 0)
df8 = pd.read_csv('최종_변환_산업경제_총사업체증감률_2013-2018.csv', index_col = 0)
df9 = pd.read_csv('최종_변환_산업경제_총종사자증감률_2013-2018.csv', index_col = 0)
df14 = pd.read_csv('최종_변환_산업경제_고차산업종사자비율_2013-2018.csv', index_col=0)

# 물리환경 부문 
df1 = pd.read_csv('최종_변환_물리환경_노후주택_신규주택 비율.csv', index_col = 0)
df2 = pd.read_csv('최종_변환_물리환경_부동산공시지가_2013-2021.csv', index_col = 0) 
```

```python
# 인구사회_인구증감률 (9년간)
df10 = df10[['dong','인구상대감소지수_std']]
df10.columns=['dong','std_people_desc']
df10
```

```python
# 산업경제_천명당총종사자수 (2018)
df7 = df7[['dong', 'std_emp_R-LQ2018']]
df7
```

```python
# 물리환경_부동산 공시지가 (2021)
df2 = df2[['dong', 'std_land_R-LQ2021']]
df2
```

```python
# 위 데이터 결합 

from functools import reduce

data_frames = [df10, df13, df12, df9, df5, df3, df7, df14, df8, df1, df2]

df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['dong']), data_frames)
df_merged 
```

```python
# 컬럼명 변경

df_merged.rename(columns = 
                {'std_people_desc':'인구변화율',
                'std_senior_child_rate':'노령화지수',
                '천명당기초생활수급자수_std':'기초생활수급(천)',
                'std_employee_decr_idx(13-18)':'총종사자증감률',
                '제조업체증감율_상대감소지수_std':'제조업체증감률',
                '도소매업체증감율_상대감소지수_std':'도소매업체증감률',
                'std_emp_R-LQ2018':'종사자수(천)',
                'std_non_h_rate_2018':'고차산업종사자수비율',
                'std_business_decr_idx(13-18)':'총사업체증감률',
                'std_over_20_years_rate':'노후주택비율',
                'std_non_within_5_years_rate':'신규주택비율',
                'std_land_R-LQ2021':'부동산공시지가'}, inplace=True)
df_merged
```

```python
# 복합쇠퇴지수 계산 ('표준화*가중치'의 합)

df_merged2 = df_merged.copy()

df_merged2['인구사회쇠퇴지수'] = (df_merged2['인구변화율']*9.246193836) + (df_merged2['노령화지수']*9.110038371) + (df_merged2['기초생활수급(천)']*8.35499443)

df_merged2['산업경제쇠퇴지수'] = (df_merged2['총종사자증감률']*5.656640673) + (df_merged2['총사업체증감률']*9.865082312) + (df_merged2['제조업체증감률']*7.389528407) + (df_merged2['도소매업체증감률']*9.097660602) + (df_merged2['종사자수(천)']*8.144572348) + (df_merged2['고차산업종사자수비율']*6.498329001)

df_merged2['물리환경쇠퇴지수'] = (df_merged2['노후주택비율']*7.637083797) + (df_merged2['신규주택비율']*8.862482981) + (df_merged2['부동산공시지가']*10.13739324)

df_merged2['복합쇠퇴지수'] = (df_merged2['인구변화율']*9.246193836) + (df_merged2['노령화지수']*9.110038371) + (df_merged2['기초생활수급(천)']*8.35499443) + (df_merged2['총종사자증감률']*5.656640673) + (df_merged2['총사업체증감률']*9.865082312) + (df_merged2['제조업체증감률']*7.389528407) + (df_merged2['도소매업체증감률']*9.097660602) + (df_merged2['종사자수(천)']*8.144572348) + (df_merged2['고차산업종사자수비율']*6.498329001) + (df_merged2['노후주택비율']*7.637083797) + (df_merged2['신규주택비율']*8.862482981) + (df_merged2['부동산공시지가']*10.13739324)

df_merged2
```

```python
# 복합쇠퇴지수 순위 산정

df_merged2['순위(복합)'] = df_merged2['복합쇠퇴지수'].rank(ascending=False) 
df_merged2['순위(인구사회)'] = df_merged2['인구사회쇠퇴지수'].rank(ascending=False) 
df_merged2['순위(산업경제)'] = df_merged2['산업경제쇠퇴지수'].rank(ascending=False) 
df_merged2['순위(물리환경)'] = df_merged2['물리환경쇠퇴지수'].rank(ascending=False) 

df_merged2
```

```python
# 필요 컬럼만 추출

df_final = df_merged2[['dong','인구사회쇠퇴지수','산업경제쇠퇴지수','물리환경쇠퇴지수','복합쇠퇴지수','순위(복합)','순위(인구사회)','순위(산업경제)','순위(물리환경)']]
df_final
```

```python
df_final.to_csv('최종_인구사회_산업경제_물리환경_복합_쇠퇴지수_순위포함.csv')
```
