```python
# 주민등록인구 : resident registration population (rereg_population)
# 2013 ~ 2022 월별 데이터

rereg_population = pd.read_excel('5_연수구_성연령별_주민등록인구.xlsx')
rereg_population.head()
```

# 전처리
- 해당 연도의 값은 매년 말월(12월)을 기준으로 함

```python
rereg_population2 = rereg_population.copy()
```

```python
# 날짜값을 문자열로 변경
rereg_population2['base_date'] = rereg_population2['base_date'].astype(str)
```

```python
# 12월만 추출하여 데이터프레임 형성

date_list = list(rereg_population2['base_date'])
x_december = []

for i in range(len(date_list)):
    if rereg_population2['base_date'][i][-2:] == '12':
        # i 는 12월 값의 인덱스 번호
        x_december.append(i)

# x_december

temp = {}
for i in x_december:
    temp[i] = rereg_population2.loc[i]

rereg_population_12 = pd.DataFrame(temp).T
```

```python
### 12월 값을 기준으로 주민등록인구수 분석

rereg_population_12.reset_index(inplace = True)
rereg_population_12 = rereg_population_12.drop(columns = 'index')
rereg_population_12.head()
```

```python
# 연도만 추출
date_list = list(rereg_population_12['base_date'])
year = ['2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

for i in range(len(date_list)):
    for j in range(len(year)):
        if date_list[i][:-2] == year[j]:
            date_list[i] = year[j]
            
rereg_population_12['base_date'] = date_list
```

```python
rereg_population_12['base_date'].unique()
```

```python
# '인천광역시 연수구 ' 라는 값 존재
# 1동,2동 등으로 구분
rereg_population_12['emd_nm'].unique()
```

```python
# 읍면동 명칭 수정
rereg_population_12['emd_nm'] = rereg_population_12['emd_nm'].apply(lambda x :'옥련동' if x in ['인천광역시 연수구 옥련1동',
                                                                                         '인천광역시 연수구 옥련2동'] else x )

rereg_population_12['emd_nm'] = rereg_population_12['emd_nm'].apply(lambda x :'연수동' if x in ['인천광역시 연수구 연수1동',
                                                                                         '인천광역시 연수구 연수2동',
                                                                                         '인천광역시 연수구 연수3동'] else x )

rereg_population_12['emd_nm'] = rereg_population_12['emd_nm'].apply(lambda x :'동춘동' if x in ['인천광역시 연수구 동춘1동', 
                                                                                         '인천광역시 연수구 동춘2동', 
                                                                                         '인천광역시 연수구 동춘3동'] else x )

rereg_population_12['emd_nm'] = rereg_population_12['emd_nm'].apply(lambda x :'송도동' if x in ['인천광역시 연수구 송도1동',
                                                                                         '인천광역시 연수구 송도2동',
                                                                                         '인천광역시 연수구 송도3동',
                                                                                         '인천광역시 연수구 송도4동',
                                                                                         '인천광역시 연수구 송도5동'] else x )

rereg_population_12['emd_nm'] = rereg_population_12['emd_nm'].apply(lambda x :'선학동' if x in ['인천광역시 연수구 선학동'] else x )

rereg_population_12['emd_nm'] = rereg_population_12['emd_nm'].apply(lambda x :'청학동' if x in ['인천광역시 연수구 청학동'] else x )
```

# 최종 : 주민등록인구_전체

```python
# 18 ~ 21 뽑기
rereg_pop_1822 = rereg_population_12[rereg_population_12['base_date'] >= '2018']
rereg_pop_1821 = rereg_pop_1822[rereg_pop_1822['base_date'] < '2022']
rereg_pop_1821
```

```python
rereg_pop_1821['total_pop'] = rereg_pop_1821['m_total_pop'] + rereg_pop_1821['w_total_pop']
rereg_pop_1821['total_0g_pop'] = rereg_pop_1821['m_0g_pop'] + rereg_pop_1821['w_0g_pop']
rereg_pop_1821['total_10g_pop'] = rereg_pop_1821['m_10g_pop'] + rereg_pop_1821['w_10g_pop']
rereg_pop_1821['total_20g_pop'] = rereg_pop_1821['m_20g_pop'] + rereg_pop_1821['w_20g_pop']
rereg_pop_1821['total_30g_pop'] = rereg_pop_1821['m_30g_pop'] + rereg_pop_1821['w_30g_pop']
rereg_pop_1821['total_40g_pop'] = rereg_pop_1821['m_40g_pop'] + rereg_pop_1821['w_40g_pop']
rereg_pop_1821['total_50g_pop'] = rereg_pop_1821['m_50g_pop'] + rereg_pop_1821['w_50g_pop']
rereg_pop_1821['total_60g_pop'] = rereg_pop_1821['m_60g_pop'] + rereg_pop_1821['w_60g_pop']
rereg_pop_1821['total_70g_pop'] = rereg_pop_1821['m_70g_pop'] + rereg_pop_1821['w_70g_pop']
rereg_pop_1821['total_80g_pop'] = rereg_pop_1821['m_80g_pop'] + rereg_pop_1821['w_80g_pop']
rereg_pop_1821['total_90g_pop'] = rereg_pop_1821['m_90g_pop'] + rereg_pop_1821['w_90g_pop']
rereg_pop_1821['total_100g_pop'] = rereg_pop_1821['m_100g_pop'] + rereg_pop_1821['w_100g_pop']
```

```python
rereg_pop_1821.columns
```

```python
rereg_pop_1821 = rereg_pop_1821.drop(columns = ['m_total_pop', 'w_total_pop','m_0g_pop', 'm_10g_pop','m_20g_pop', 'm_30g_pop', 'm_40g_pop', 'm_50g_pop', 
                                                      'm_60g_pop', 'm_70g_pop', 'm_80g_pop', 'm_90g_pop', 'm_100g_pop','w_0g_pop', 
                                                      'w_10g_pop', 'w_20g_pop', 'w_30g_pop', 'w_40g_pop','w_50g_pop', 'w_60g_pop', 
                                                      'w_70g_pop', 'w_80g_pop', 'w_90g_pop','w_100g_pop'])
```

```python
rereg_pop_1821.reset_index(inplace = True)
rereg_pop_1821 = rereg_pop_1821.drop(columns = 'index')
```

```python
rereg_pop_1821 = rereg_pop_1821.groupby(['base_date', 'emd_nm']).sum()
rereg_pop_1821.reset_index(inplace = True)
rereg_pop_1821.head()
```

# 최종 : 주민등록인구_동별_계층

```python
# '인천광역시 연수구 ' 삭제를 위해 인덱스로 'emd_nm' 설정
dong_rereg_pop_1821 = rereg_pop_1821.set_index('emd_nm')

# 인덱스명이 '인천광역시 연수구 '인 행 삭제
dong_rereg_pop_1821.drop('인천광역시 연수구 ', axis = 0, inplace = True)

# 인덱스 되돌리기
dong_rereg_pop_1821.reset_index(inplace = True)

dong_rereg_pop_1821.head()
```

```python
rereg_standard = dong_rereg_pop_1821.copy()
```

```python
# 아동청소년층
rereg_standard['total_youth_pop'] = rereg_standard['total_0g_pop'] + rereg_standard['total_10g_pop']

# 청장년층
rereg_standard['total_adult_pop'] = (rereg_standard['total_20g_pop'] + rereg_standard['total_30g_pop'] 
                                     + rereg_standard['total_40g_pop'] + rereg_standard['total_50g_pop'])
# 노년층
rereg_standard['total_senior_pop'] = (rereg_standard['total_60g_pop'] + rereg_standard['total_70g_pop']
                                + rereg_standard['total_80g_pop'] + rereg_standard['total_90g_pop'] + rereg_standard['total_100g_pop'])
```

```python
rereg_standard = rereg_standard.drop(columns = ['total_0g_pop', 'total_10g_pop', 'total_20g_pop', 'total_30g_pop', 'total_40g_pop', 'total_50g_pop',
                              'total_60g_pop', 'total_70g_pop', 'total_80g_pop', 'total_90g_pop', 'total_100g_pop'])
```

```python
rereg_standard = rereg_standard.groupby(['base_date', 'emd_nm']).sum()
rereg_standard.reset_index(inplace = True)
rereg_standard.head()
```

```python
rereg_standard['senior_rate'] = (rereg_standard['total_senior_pop'] / rereg_standard['total_pop']) * 100
rereg_standard['youth_rate'] = (rereg_standard['total_youth_pop'] / rereg_standard['total_pop']) * 100
rereg_standard['adult_rate'] = (rereg_standard['total_adult_pop'] / rereg_standard['total_pop']) * 100
```

```python
rereg_standard.head()
```

```python
# 저장
rereg_standard.to_csv("최종_주민등록인구_연령구분_1821.csv", encoding = 'utf-8', index = False)
```

# 최종 : 주민등록인구_연령구분_신도시_원도심

```python
# 원도심
rereg_standard_old = rereg_standard[rereg_standard['emd_nm'] != '송도동']
rereg_standard_old = rereg_standard_old.drop(columns = ['senior_rate', 'youth_rate', 'adult_rate'])
rereg_standard_old = rereg_standard_old.groupby('base_date').sum()
rereg_standard_old.reset_index(inplace = True)
rereg_standard_old.insert(1, 'center', '원도심')
rereg_standard_old
```

```python
rereg_standard_old['senior_rate'] = (rereg_standard_old['total_senior_pop'] / rereg_standard_old['total_pop']) * 100
rereg_standard_old['youth_rate'] = (rereg_standard_old['total_youth_pop'] / rereg_standard_old['total_pop']) * 100
rereg_standard_old['adult_rate'] = (rereg_standard_old['total_adult_pop'] / rereg_standard_old['total_pop']) * 100
```

```python
rereg_standard_old
```

```python
# 신도심
rereg_standard_new = rereg_standard[rereg_standard['emd_nm'] == '송도동']
rereg_standard_new= rereg_standard_new.groupby('base_date').sum()
rereg_standard_new.reset_index(inplace = True)
rereg_standard_new.insert(1, 'center', '신도시')
rereg_standard_new
```

```python
# 합치기
rereg_standard_old_new = pd.concat([rereg_standard_new, rereg_standard_old])
rereg_standard_old_new = rereg_standard_old_new.sort_values(by = rereg_standard_old_new.columns[0])
rereg_standard_old_new.reset_index(inplace = True)
rereg_standard_old_new = rereg_standard_old_new.drop(columns = 'index')
rereg_standard_old_new
```

```python
# 저장
rereg_standard_old_new.to_csv("최종_주민등록인구_신도시_원도심.csv", encoding = 'utf-8', index = False)
```
