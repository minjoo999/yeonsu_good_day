```python
## 영업구분 새로운 값으로 변경
#    '폐업', '폐업처리', '직권폐업', '임시소매기간만료' >> '폐업'
#    '직권말소', '말소', '허가취소', '등록취소', '직권취소', '지정취소' >> '삭제'
#    '영업중', '정상', '영업', '정상영업', '영업재개' >> '영업'
#    '휴업', '휴업처리', '영업정지' >> '휴업'

# 값 변경
Business2['open_type'] = Business2['open_type'].replace(['폐업', '폐업처리', '직권폐업', '임시소매기간만료'], '폐업')
Business2['open_type'] = Business2['open_type'].replace(['직권말소', '말소', '허가취소', '등록취소', '직권취소', '지정취소'], '삭제')
Business2['open_type'] = Business2['open_type'].replace(['영업중', '정상', '영업', '정상영업', '영업재개'], '영업')
Business2['open_type'] = Business2['open_type'].replace(['휴업', '휴업처리', '영업정지'], '휴업')

Business2['open_type'].unique()
```

```python
Business2['open_type'].value_counts()
```

```python
Business2.reset_index(inplace = True)
Business2 = Business2.drop(columns = 'index')
```

```python
# 년월일로 되어있는 날짜데이터를 년도로 변경

# license_date : 영업허가 일자

# 자료형 변경
Business2['license_date'] = Business2['license_date'].astype(str)

license_temp = []

for i in range(len(Business2)):
    license_temp.append(Business2.loc[i,'license_date'][:4])

Business2['license_date'] = license_temp
```

```python
# close_date : 폐업 일자

# 폐업하지 않은 가게도 있기 때문에, close_date 결측값 임의로 설정
Business2['close_date'] = Business2['close_date'].fillna('00000000')

# 자료형 변경
Business2['close_date'] = Business2['close_date'].astype(str)

close_temp = []
for i in range(len(Business2)):
    close_temp.append(Business2.loc[i,'close_date'][:4])

Business2['close_date'] = close_temp
```

```python
# store_type 변경
# 건강, 동물, 문화, 생활, 식품, 기타 
health = ['병원', '의원', '산후조리원', '안전상비의약품 판매업소', '약국', '안경업']
animal = ['동물미용업', '동물판매업', '동물약국', '동물병원']
culture = ['인터넷컴퓨터게임시설제공업', '일반게임제공업', '청소년게임제공업', '노래연습장업', '숙박업']
life = ['미용업', '이용업', '통신판매업', '당구장업']
food = [' 건강기능식품일반판매업', '축산판매업', '즉석판매제조가공업', '단란주점영업', '유흥주점영업', '일반음식점', '휴게음식점']
etc = ['담배소매업']
```

```python
type_list = []

for i in Business2['store_type']:
    if i in health:
        type_list.append('건강')
    elif i in animal:
        type_list.append('동물')
    elif i in culture:
        type_list.append('문화')
    elif i  in life:
        type_list.append('생활')
    elif i  in food:
        type_list.append('식품')
    else:
        type_list.append('기타')
```

```python
Business2['store_type'] = type_list
```

# 최종 : 상권정보_연도별_창업_폐업_1821.csv

```python
Business3 = Business2.copy()
```

```python
Business3_open = Business3[Business3['license_date'] >= '2018']
Business3_open = Business3_open[Business3_open['license_date'] <= '2021']
Business3_open.reset_index(inplace = True)
Business3_open = Business3_open.drop(columns = ['index', 'close_date'])
Business3_open['open_cnt'] = 1
Business3_open = Business3_open.groupby(['emd_nm', 'license_date', 'store_type']).sum()
Business3_open.reset_index(inplace = True)
Business3_open = Business3_open.rename(columns = {'license_date': 'base_date'})
Business3_open
```

```python
Business3_open.isnull().sum()
```

```python
Business3_close = Business3[Business3['close_date'] >= '2018']
Business3_close = Business3_close[Business3_close['close_date'] <= '2021']
Business3_close.reset_index(inplace = True)
Business3_close = Business3_close.drop(columns = ['index', 'license_date'])
Business3_close['close_cnt'] = 1
Business3_close = Business3_close.groupby(['emd_nm', 'close_date', 'store_type']).sum()
Business3_close.reset_index(inplace = True)
Business3_close = Business3_close.rename(columns = {'close_date': 'base_date'})
Business3_close
```

```python
Business3_close.isnull().sum()
```

```python
# 저장
Business3_open.to_csv("최종_상권정보_창업_1821.csv", encoding = 'utf-8', index = False)
Business3_close.to_csv("최종_상권정보_폐업_1821.csv", encoding = 'utf-8', index = False)
```

```python
Business3_open_old = Business3_open[Business3_open['emd_nm'] != '송도동']
Business3_open_old = Business3_open_old.drop(columns = 'emd_nm')
Business3_open_old = Business3_open_old.groupby(['base_date', 'store_type']).sum()
Business3_open_old.reset_index(inplace = True)
Business3_open_old.insert(1, 'center','원도심')
Business3_open_old
```

```python
Business3_open_new = Business3_open[Business3_open['emd_nm'] == '송도동']
Business3_open_new = Business3_open_new.drop(columns = 'emd_nm')
Business3_open_new = Business3_open_new.groupby(['base_date', 'store_type']).sum()
Business3_open_new.reset_index(inplace = True)
Business3_open_new.insert(1, 'center','신도시')
Business3_open_new
```

```python
Business3_open_old_new = pd.concat([Business3_open_new, Business3_open_old])
Business3_open_old_new = Business3_open_old_new.sort_values(by = [Business3_open_old_new.columns[0],Business3_open_old_new.columns[1]] )
Business3_open_old_new.reset_index(inplace = True)
Business3_open_old_new = Business3_open_old_new.drop(columns = 'index')
Business3_open_old_new.head()
```

```python
Business3_close_old = Business3_close[Business3_close['emd_nm'] != '송도동']
Business3_close_old = Business3_close_old.drop(columns = 'emd_nm')
Business3_close_old = Business3_close_old.groupby(['base_date', 'store_type']).sum()
Business3_close_old.reset_index(inplace = True)
Business3_close_old.insert(1, 'center','원도심')
Business3_close_old
```

```python
Business3_close_new = Business3_close[Business3_close['emd_nm'] == '송도동']
Business3_close_new = Business3_close_new.drop(columns = 'emd_nm')
Business3_close_new = Business3_close_new.groupby(['base_date', 'store_type']).sum()
Business3_close_new.reset_index(inplace = True)
Business3_close_new.insert(1, 'center','신도심')
Business3_close_new
```

```python
Business3_close_old_new = pd.concat([Business3_close_new, Business3_close_old])
Business3_close_old_new = Business3_close_old_new.sort_values(by = [Business3_close_old_new.columns[0],Business3_close_old_new.columns[1]] )
Business3_close_old_new.reset_index(inplace = True)
Business3_close_old_new = Business3_close_old_new.drop(columns = 'index')
Business3_close_old_new
```
